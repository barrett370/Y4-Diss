```julia
t = 0.6 
c = BezierCurve([Point(0,0),Point(2,5),Point(5,4)])
c1 = [ c[1:i](t) for i in 1:length(c)]
c2 = [ c[i:length(c)](1-t) for i in length(c):-1:1]
```

```julia
p = plot_curve(c,100)
ps_x, ps_y = [], []
for x in range(0, 1, step = 1 / 100)
    C = c1(x)
    append!(ps_x, C.x)
    append!(ps_y, C.y)
end
p = plot!(p,(ps_x, ps_y),linewidth=3,legend=false)
```
